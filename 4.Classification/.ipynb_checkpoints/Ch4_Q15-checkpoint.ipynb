#### **(D)** Write a function `power3`, that returns the resutl x**a as a python object.

```python
def power3(x,a):
    result = x**a
    return result
```

#### **(E)** Create a plot of $f(x) = x^2$ using `power3` function

```python
x = np.arange(1,11)
y = power3(x,2)

fig,ax = plt.subplots(figsize = (4,4))
ax.plot(x,y, marker= 'o', linestyle = '-', label = r'$y = x^2$')
ax.set_xlabel ('x', fontsize=10)
ax.set_ylabel (r'$y = x^2$', fontsize = 10)
ax.set_title(r'$y = x^2$', fontsize = 12)
ax.legend()

#ax.set_xscale('log')  
#ax.set_yscale('log') 
#ax.set_xscale('log'); ax.set_yscale('log')  

```

#### **(F)** Create a function, `PlotPower()`, that creates a plot of x against x**a 

```python
def PlotPower(x,a):
    y = power3(x,a)
    fig,ax = plt.subplots(figsize = (4,4))
    ax.plot(x,y, marker= 'o', linestyle = '-', label = fr'$y = x^{a}$')
    ax.set_xlabel ('x', fontsize=10)
    ax.set_ylabel (fr'$y = x^{a}$', fontsize = 10)
    ax.set_title(fr'$y = x^{a}$', fontsize = 12)
    ax.legend()

PlotPower (np.arange (1, 11) , 3)
```
