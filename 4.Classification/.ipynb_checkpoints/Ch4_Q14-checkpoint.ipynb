```python
import os
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
from ISLP import load_data
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
auto = load_data('Auto')
auto.head()
```

### **(A)** 
Create a binary variable `mpg01`, that contains a 1 if `mpg` contains a value above its median, and a 0 if `mpg` contains a value below its median

```python
#weekly['Direction_bin'] = weekly['Direction'].map({'Up': 1, 'Down': 0})

mpg_median = np.median(auto['mpg'])
auto['mpg01'] = (auto['mpg'] > mpg_median).astype(int)
# or 
#auto['mpg01'] = auto['mpg'].apply(lambda x: 1 if x > mpg_median else 0)
```

### **(B)** Association between `mpg01` and the other features

```python
auto_df = pd.DataFrame(auto)
pd.plotting.scatter_matrix(auto_df, figsize=(16,16))
```

```python
# weight, acceleration, horsepower
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (8,3))
# box plot for weights
ax1.boxplot([auto_df.loc[auto_df['mpg01']==0, 'weight'],
             auto_df.loc[auto_df['mpg01']==1, 'weight']], 
            tick_labels = ['mpg01-0','mpg01-1'])
ax1.set_xlabel("mpg01")
ax1.set_ylabel("weight")

# box plot for acceleration
ax2.boxplot([auto_df.loc[auto_df['mpg01']==0, 'acceleration'],
             auto_df.loc[auto_df['mpg01']==1, 'acceleration']], 
            tick_labels = ['mpg01-0','mpg01-1'])
ax2.set_xlabel("mpg01")
ax2.set_ylabel("acceleration")

# box plot for horsepower
ax3.boxplot([auto_df.loc[auto_df['mpg01']==0, 'horsepower'],
             auto_df.loc[auto_df['mpg01']==1, 'horsepower']], 
            tick_labels = ['mpg01-0','mpg01-1'])
ax3.set_xlabel("mpg01")
ax3.set_ylabel("horsepower")


# Adjust layout
plt.tight_layout()
plt.show()
```

### **(C)** split the data into a training set and a test set

```python
# Select only the required predictors
predictors = auto_df[['weight', 'horsepower', 'acceleration']]
y = auto_df['mpg01']
X_train, X_test, y_train, y_test = train_test_split(predictors, y, test_size = 0.3, random_state = 42, stratify = y)
```

### **(D)** Perform LDA


```python
lda = LDA(store_covariance=True)
lda_fit = lda.fit(X_train, y_train)

# evaluate with the test data set
y_predict = lda_fit.predict(X_test)
print("The error rate of the LDA model is {0:.2f}%".format(100*np.mean(y_predict!=y_test)))
```

### **(E)** Perform QDA

```python
qda = QDA(store_covariance=True)
qda_fit = qda.fit(X_test, y_test)

# evaluate the test dataset
y_predict_qda = qda_fit.predict(X_test)
print("The error rate of the QDA model is {0: 0.2f}%".format(100*np.mean(y_predict_qda != y_test)))
```

### **(F)** Perform logistic regression

```python
X_log = sm.add_constant(X_train)
log_model = sm.GLM(y_train, X_log, family= sm.families.Binomial()).fit()
print(log_model.summary())

X_test_trnsfrm = sm.add_constant(X_test)
y_log_predict = log_model.predict(X_test_trnsfrm)

y_log_predict_true = (y_log_predict>0.5).astype(int)
print(confusion_matrix(y_log_predict_true, y_test))
(50+53)/np.size(y_log_predict_true), np.mean(y_log_predict_true == y_test)
print("Error rate is ", 1 - (50+53)/np.size(y_log_predict))
```

### **(G)** Perform naive Bayes

```python
NB = GaussianNB()
NB_fit = NB.fit(X_train, y_train)

# evaluate with the test data set
y_NB_predict = NB_fit.predict(X_test)
print("The error rate of the LDA model is {0:.2f}%".format(100*np.mean(y_NB_predict!=y_test)))
```

### **(H)** Perform KNN with multiple levels of K


```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  
X_test_scaled = scaler.transform(X_test)  # transform only applies the scaling

k_values = [1, 5, 10, 15, 20, 25]
test_errors = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k) # train
    knn.fit(X_train_scaled, y_train)
    
    # prediction
    y_pred = knn.predict(X_test_scaled)
    
    # test error
    error = 1 - accuracy_score(y_test, y_pred)
    test_errors.append(error)
    print(f"K={k}: Test Error = {error:.4f} ({error*100:.2f}%)")

# plot test errors for different Ks
plt.figure(figsize=(8, 5))
plt.plot(k_values, test_errors, marker='o', linestyle='-')
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Test Error Rate")
plt.title("KNN Test Error vs. K")
plt.xticks(k_values)
plt.grid(True)
plt.show()

best_k = k_values[np.argmin(test_errors)]
print(f"Best K: {best_k} with Test Error = {min(test_errors):.4f} ({min(test_errors)*100:.2f}%)")
```
