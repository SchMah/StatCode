This question should be answered using the Carseats data set.

**(a)** Fit a multiple regression model to predict Sales using Price,
Urban, and US.

**(b)** Provide an interpretation of each coefficient in the model. Be
careful—some of the variables in the model are qualitative!

**(c)** Write out the model in equation form, being careful to handle
the qualitative variables properly.

**(d)** For which of the predictors can you reject the null hypothesis
H0 : βj = 0?

**(e)** On the basis of your response to the previous question, fit a
smaller model that only uses the predictors for which there is
evidence of association with the outcome.

**(f)** How well do the models in (a) and (e) fit the data?

**(g)** Using the model from (e), obtain 95 % confidence intervals for
the coefficient(s).

**(h)** Is there evidence of outliers or high leverage observations in the
model from (e)?

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 
from ISLP import load_data
```

```python
# (a) sales~price+urban+us
car_data = load_data('Carseats')
print(car_data.head())
y = car_data['Sales']
car_data['Urban'] = car_data['Urban'].map({'Yes': 1, 'No': 0})
car_data['US'] = car_data['US'].map({'Yes': 1, 'No': 0})
X = sm.add_constant(car_data[['Price','Urban','US']])
results = sm.OLS(y,X).fit()
results.summary()
```

**(b)** Provide an interpretation of each coefficient in the model. Be
careful—some of the variables in the model are qualitative!

The price has a negative coef and statistically significant (p < 0.05), meaning that as the car price increases, the sale decreases. 
The urban predictor has also a negative slope coef but not statistically significant.
The US predictor has a positive relationship with sales at US=1. This shows the sales increases in the US market (US = 1)

**(c)** Write out the model in equation form, being careful to handle
the qualitative variables properly.

$Sales \sim \beta_{\text{Price}} X_{\text{price}} + \beta_{\text{US}} X_{\text{US}} + \beta_{\text{Urban}} X_{\text{Urban}} + \epsilon$.

**(d)** For which of the predictors can you reject the null hypothesis
H0 : βj = 0?
Price and US

**(e)** On the basis of your response to the previous question, fit a
smaller model that only uses the predictors for which there is
evidence of association with the outcome.


```python
# (e) Use only Price and US as predictors Sales ~ Price + US
X_two = sm.add_constant(car_data[['Price','US']])
results_two = sm.OLS(y, X_two).fit()
results_two.summary()
```

```python
# (f) How well do the models in (a) and (e) fit the data?
model_eval = sm.stats.anova_lm(results, results_two)
print(model_eval)
```

Neither of the models fit the data well. The R squared is the same for both models. 

**(g)** Using the model from (e), obtain 95 % confidence intervals for
the coefficient(s).

95% confidence interval for the Price predictor is [-0.06 -0.04] and for the US predictor is [0.69 1.70]

```python
#(h) Diagnostic plots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(8,3))

# subplot (1,1) : fitted vs residual values
ax1.scatter(results.fittedvalues, results.resid, alpha = 0.3)
ax1.set_xlabel('fitted values')
ax1.set_ylabel('residuals')
ax1.axhline(0, c='k', ls='--')

infl = results.get_influence()
ax2.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag,alpha = 0.3)
ax2.set_xlabel('Index')
ax2.set_ylabel('Leverage')
np.argmax(infl.hat_matrix_diag)
# second model with fewer predictors
ax3.scatter(results_two.fittedvalues, results_two.resid, alpha = 0.3)
ax3.set_xlabel('fitted values')
ax3.set_ylabel('residuals')
ax3.axhline(0, c='k', ls='--')

infl = results_two.get_influence()
ax4.scatter(np.arange(X_two.shape[0]), infl.hat_matrix_diag, alpha = 0.3)
ax4.set_xlabel('Index')
ax4.set_ylabel('Leverage')
np.argmax(infl.hat_matrix_diag)
```
