```python
import os
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
np.random.seed(1)
```

```python
# (a), (b), (c), and (d)
x = np.random.randn(100)
eps = 0.25*np.random.randn(100)
Y = -1 + 0.5 * x + eps
print(len(Y))
fig, ax = plt.subplots(figsize = (4,4))
ax.scatter(x,Y)
print("The length Y is", len(Y), "and beta0 and beta1 in the model are -1 and 0.5, respectively")
```

```python
# (e) and (f) fit a least square model and draw the regression line
x = sm.add_constant(x)
results = sm.OLS(Y,x).fit()
print(results.summary())
y_predict = results.predict(x)
fig, ax = plt.subplots(figsize = (4,4))
ax.scatter(x[:,1],Y, alpha = 0.3)
# plot the regression line
ax.plot(x[:,1], y_predict, label = "Least Square Regression Line", color = 'red')
# plot the population regression line
b_intercept = -1 
b_slope = 0.50  
y_pop_line = b_intercept + b_slope * x[:,1]
ax.plot(x[:,1], y_pop_line, label="Population Regression Line", color="green", linestyle="dashed")
ax.legend()
```

```python
# (g) fit a polynomial regression line 
x_poly = x[:,1]**2
x_new = np.concatenate((x[:,1].reshape(-1,1), np.reshape(x_poly, (len(x), 1))), axis = 1)
X_new = sm.add_constant(x_new)
results = sm.OLS(Y,X_new).fit()
print(results.summary())
```
