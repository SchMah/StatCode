```python
import os
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ISLP import load_data
```

```python
boston = load_data('Boston')
boston.head()
#fig = pd.plotting.scatter_matrix(boston, figsize= (18,18))
#np.shape(boston)[1]
results_list = [] 
predictors = boston.columns.drop(['crim'])
for i in list(predictors):
    x = boston[i]
    X = sm.add_constant(x)
    y = boston['crim']
    results = sm.OLS(y, X).fit()
#    print("The statistical significant between crime rate and", i, "is", results.pvalues)
    print("{0:8s}: [beta_0, beta_1] = [{1:5.2f},{2:6.2f}]:    p-value= {3:10.6f}".format(i, results.params.iloc[0], results.params.iloc[1], results.pvalues.iloc[1]))
    results_list.append({
        'Predictor': i,
        'Beta_0': results.params.iloc[0],
        'Beta_1': results.params.iloc[1],
        'P_Value': results.pvalues.iloc[1]
    })
    
    
```

```python
#(b) multiple regression model
x = boston[boston.columns.drop(['crim'])]
X = sm.add_constant(x)
y = boston['crim']
results_mul = sm.OLS(y,X).fit()
print(results_mul.summary())
```

```python
#(c)
results_df = pd.DataFrame(results_list)

fig, ax = plt.subplots(figsize = (6,6))
for i in list(predictors):
    x_uni = results_df.loc[results_df['Predictor'] == i, 'Beta_1'].values[0] 
    y_multi = results_mul.params[i]  
    ax.scatter(x_uni, y_multi, label=i)

ax.set_xlabel('Beta_1 (Univariate)')
ax.set_ylabel('Beta_1 (Multivariate)')
ax.set_title('Comparison of Univariate and Multivariate Coefficients')
ax.axhline(0, color='black', linewidth=1)  
ax.axvline(0, color='black', linewidth=1)
ax.legend(loc='best')

plt.show()

```

```python
#(d)
predictors = boston.columns.drop(['crim'])
for i in list(predictors):
    x = boston[i]
    x_new = pd.DataFrame({'X':x,
                     'X2': x**2,
                     'X3': x**3})
    X = sm.add_constant(x_new)
    y = boston['crim']
    results_quad = sm.OLS(y,X).fit()
    print("{0:8s}: [beta_0, beta_1, beta_2, beta_3] = [{1:5.2f},{2:6.2f},{3:5.2f},{4:5.2f}]: p-value= {5:10.6f}, {6:10.6f}, {7:10.6f}".format(
        i, 
        results_quad.params.iloc[0],  # Beta_0
        results_quad.params.iloc[1],  # Beta_1
        results_quad.params.iloc[2],  # Beta_2
        results_quad.params.iloc[3],  # Beta_3
        results_quad.pvalues.iloc[1],  # P-value for Beta_1
        results_quad.pvalues.iloc[2],  # P-value for Beta_2
        results_quad.pvalues.iloc[3]  # P-value for Beta_3
    ))
  

    
```
