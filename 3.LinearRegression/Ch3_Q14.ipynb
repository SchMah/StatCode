```python
import os
import numpy as np 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rng = np.random.default_rng(10)
x1 = rng.uniform(0,1,size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)
```

$ y = 2 + 2 * X_{\text{1}} + 0.3 * X_{\text{2}} + eps$

$beta_{\text{0}} = 2, beta_{\text{1}} = 2, beta_{\text{2}} = 0.3$

```python
#(b) the correlation between x1 and x2
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize = (6,2))
ax1.scatter(x1,x2,alpha=0.3)
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
corr_coef = np.corrcoef(x1,x2)
print("The correlation between x1 and x2 is", corr_coef[1,0])

ax2.scatter(x1,y,alpha=0.3)
ax2.set_xlabel('x1')
ax2.set_ylabel('y')

ax3.scatter(x2,y,alpha=0.3)
ax3.set_xlabel('x2')
ax3.set_ylabel('y')
```

```python
# (c) fit a least square regression
x = np.concatenate((np.reshape(x1,(len(x1),1)),np.reshape(x2,(len(x2),1))),axis = 1)
X = sm.add_constant(x)
results = sm.OLS(y,X).fit()
print(results.summary())
```

The intercept is significant and close to true $ \beta_0 $. $ \beta_1 $ is also significant and we can reject the null hypothesis that $ \beta_1 $ is 0. However, $ \beta_2 $ is not significant. Thus, the null hypothesis cannot be rejected. 


```python
# (d) and (e) fit a least square regression to predict y using only x1 or x2
X_x1 = sm.add_constant(x1)
results_x1 = sm.OLS(y,X_x1).fit()
print(results_x1.summary())

X_x2 = sm.add_constant(x2)
results_x2 = sm.OLS(y,X_x2).fit()
print(results_x2.summary())

```

**d/e/f*** In the first model (using only x1), the coefficient is close to the true value in the main model and significant. The second model also has a significant coefficient, unlike the model considering both predictors (x1 and x2). Both predictors are independetly correlated with y and it seems only one is enough to capture the variance in data. 

```python
# (g) - adding outliers to data and repeat c-e
x1 = np.concatenate([x1, [0.1]])
x2 = np.concatenate([x2, [0.8]])
y = np.concatenate([y, [6]])

# fit the model using x1 and x2
x = np.concatenate((np.reshape(x1,(len(x1),1)),np.reshape(x2,(len(x2),1))),axis = 1)
X = sm.add_constant(x)
results = sm.OLS(y,X).fit()
print(results.summary())

# fit the model using only x1
X_x1 = sm.add_constant(x1)
results_x1 = sm.OLS(y,X_x1).fit()
print(results_x1.summary())

# fit the model using only x2
X_x2 = sm.add_constant(x2)
results_x2 = sm.OLS(y,X_x2).fit()
print(results_x2.summary())

#  diagnostic plots
fig, ax
```
