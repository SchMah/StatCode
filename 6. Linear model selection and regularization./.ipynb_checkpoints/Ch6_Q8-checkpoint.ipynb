```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, r2_score
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
```

```python
# (a & b) random number generator
rng = np.random.default_rng(42)
n = 100 # size of data
X = rng.normal(0,1, n)
eps = rng.normal(0,1,n)
Y = 1 + 2*X + 3*(X**2) + 4*(X**3) + eps
data = pd.DataFrame({'Y':Y,
                    'X' :X,
                    'X2': X**2,
                    'X3': X**3})

# (c)
X_train, X_test, Y_train, Y_test = train_test_split(data.drop('Y', axis = 1), data['Y'],
                                                    train_size = 0.7,
                                                    random_state= 42)

best_model_steps = []
# forward stepwise selection
def forward_stepwise_selection(x, y):
    model_predictors = []
    remaining_predictors = list(x.columns)

    # first fit with all predictors to get initial sigma squared
    X = sm.add_constant(x)
    fit_model = sm.OLS(y, X).fit()
    # 
    # calculate sigma squared as [RSS/(n−p )]
    RSS = fit_model.ssr
    n = x.shape[0]
    p = len(fit_model.params)
    sigma_squared = RSS/ (n - p)

    for _ in range(len(remaining_predictors)):
        best_model = None
        best_cp = np.inf
        best_predictors = None

        for predic in remaining_predictors:
            current_predictor = model_predictors + [predic]
            X_new = sm.add_constant(X[current_predictor])
            model_new = sm.OLS(y, X_new).fit()
            # calculate Cp
            RSS_new = model_new.ssr
            p_new = len(model_new.params)
            Cp_new = RSS_new/sigma_squared - (X_new.shape[0]-2*p_new)
            if Cp_new < best_cp:
                best_cp = Cp_new
                best_predictors = predic
                best_model = model_new

            
        model_predictors.append(best_predictors) 
        remaining_predictors.remove(best_predictors)

        best_model_steps.append((best_model, list(model_predictors), best_cp))

    return(best_model_steps)



models_steps = forward_stepwise_selection(X_train, Y_train)           
            
best_model_selected = min(models_steps, key = lambda x : x[2])  
best_model, best_predictors, best_cp = best_model_selected
print(best_model.summary())
print(best_predictors)
print(best_cp)

# some plots
number_predictors = []
r2 = []
MSE = []
Cp = []

for i in range(len(models_steps)):
    number_predictors.append(len(models_steps[i][1])) 
    Cp.append(models_steps[i][2])                    
    MSE.append(models_steps[i][0].mse_resid) 
    r2.append(models_steps[i][0].rsquared) 


fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (10,9))
ax1.plot(number_predictors,Cp, marker = 'o')
ax1.set_xlabel('predictors')
ax1.set_ylabel('Cp')

ax2.plot(number_predictors, MSE, marker = 'o')
ax2.set_xlabel('predictors')
ax2.set_ylabel('MSE')

ax3.plot(number_predictors, r2, marker = 'o')
ax3.set_xlabel('predictors')
ax3.set_ylabel('R2')
```

```python
# (d) Backward selection
# ----------- Data generation same as (b) -------
# -----------------------------------------------
rng = np.random.default_rng(42)
n = 100 # size of data
X = rng.normal(0,1, n)
eps = rng.normal(0,1,n)
Y = 1 + 2*X + 3*(X**2) + 4*(X**3) + eps
data = pd.DataFrame({'Y':Y,
                    'X' :X,
                    'X2': X**2,
                    'X3': X**3})

# (d)
X_train, X_test, Y_train, Y_test = train_test_split(data.drop('Y', axis = 1), data['Y'],
                                                    train_size = 0.7,
                                                    random_state= 42)

# start the model with full predictors included and then remove one after each iteration
best_model_steps = []
# backward stepwise selection
def backward_stepwise_selection(x, y):
    model_predictors = list(x.columns)
    

    # first fit with all predictors to get initial sigma squared
    X = sm.add_constant(x)
    fit_model = sm.OLS(y, X).fit()
    # 
    # calculate sigma squared as [RSS/(n−p )]
    RSS = fit_model.ssr
    n = x.shape[0]
    p = len(fit_model.params)
    sigma_squared = RSS/ (n - p)

    for _ in range(len(model_predictors)):
        best_model = None
        best_cp = np.inf
        best_predictors = None

        for predic in model_predictors:
            current_predictors = model_predictors.copy()
            current_predictors.remove(predic)
            X_new = sm.add_constant(X[current_predictors])
            model_new = sm.OLS(y, X_new).fit()
            # calculate Cp
            RSS_new = model_new.ssr
            p_new = len(model_new.params)
            Cp_new = RSS_new/sigma_squared - (X_new.shape[0]-2*p_new)
            if Cp_new < best_cp:
                best_cp = Cp_new
                best_predictors = predic
                best_model = model_new

            
        model_predictors.append(best_predictors) 

        best_model_steps.append((best_model, list(model_predictors), best_cp))

    return(best_model_steps)



models_steps = backward_stepwise_selection(X_train, Y_train)           
            
best_model_selected = min(models_steps, key = lambda x : x[2])  
best_model, best_predictors, best_cp = best_model_selected
print(best_model.summary())
print(best_predictors)
print(best_cp)

# some plots
number_predictors = []
r2 = []
MSE = []
Cp = []

for i in range(len(models_steps)):
    number_predictors.append(len(models_steps[i][1])) 
    Cp.append(models_steps[i][2])                    
    MSE.append(models_steps[i][0].mse_resid) 
    r2.append(models_steps[i][0].rsquared) 


fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (10,9))
ax1.plot(number_predictors,Cp, marker = 'o')
ax1.set_xlabel('predictors')
ax1.set_ylabel('Cp')

ax2.plot(number_predictors, MSE, marker = 'o')
ax2.set_xlabel('predictors')
ax2.set_ylabel('MSE')

ax3.plot(number_predictors, r2, marker = 'o')
ax3.set_xlabel('predictors')
ax3.set_ylabel('R2')
```

```python
# (d) Backward selection
# ----------- Data generation same as (b) -------
# -----------------------------------------------
rng = np.random.default_rng(42)
n = 100 # size of data
X = rng.normal(0,1, n)
eps = rng.normal(0,1,n)
Y = 1 + 2*X + 3*(X**2) + 4*(X**3) + 5*(X**4)+ 6*(X**5)+7*(X**6)+8*(X**7)+9*(X**8)+10*(X**9)+11*(X**10)+ eps
data = pd.DataFrame({'Y':Y,
                    'X' :X,
                    'X2': X**2,
                    'X3': X**3,
                    'X4': X**4,
                    'X5': X**5,
                    'X6': X**6,
                    'X7' :X**7,
                    'X8': X**8,
                    'X9': X**9,
                    'X10':X**10})
data_x = data.drop(columns = 'Y')
data_y = data['Y']

X_train, X_test, Y_train, Y_test = train_test_split (data_x,data_y, train_size= 0.7, random_state = 42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# apply lasso
lasso_model = LassoCV(alphas= None, cv = 5, random_state = 42, max_iter= 100000)
lasso_model.fit(X_train_scaled, Y_train)


mse_mean = np.mean(lasso_model.mse_path_, axis=1)  # average across folds
alphas = lasso_model.alphas_

plt.plot(alphas, mse_mean, marker='o')
plt.xscale('log')  # log scale is standard for regularization parameters
plt.xlabel("Alpha (lambda)- log scale")
plt.ylabel("Mean Cross-Validation MSE")
plt.show()

print("Best alpha (lambda) chosen by lasso:", lasso_model.alpha_)

print("Intercept:", lasso_model.intercept_)
print("Coefficients for X, X2, ..., X10:", lasso_model.coef_)


# Plot CV error vs alpha
fig, ax = plt.subplots(figsize=(10,5));
ax.plot(lasso_model.alphas_,mse_mean, color='b');
ax.plot(lasso_model.alpha_, np.min(mse_mean), marker='o', color='r', markersize=8);
ax.set_xscale('log')
ax.set_xlabel('Alpha (lambda) log scale');
ax.set_ylabel('MSE');
```
